```python
!pip install nbformat

import nbformat

# Read your original notebook
nb = nbformat.read("chatbot_langchain.ipynb", as_version=4)

# Remove widgets metadata from each cell
for cell in nb["cells"]:
    if "metadata" in cell:
        cell["metadata"].pop("widgets", None)

# Also remove any top-level widgets metadata if present
nb["metadata"].pop("widgets", None)

# Save a clean copy
nbformat.write(nb, "chatbot_langchain_clean.ipynb")

print("Cleaned notebook saved as chatbot_langchain_clean.ipynb")
```
